Read the boston dataset csv files provided for this assignment into Python (you can use pd.read_csv()). Combine or concatenate the DataFrames, boston_1, boston_2, boston_3, and boston_4 and assign the results to a DataFrame called combined_boston. These four datasets should be combined vertically since they have the same variable names, such that boston_1 is stacked on top of boston_2, and the result is stacked on top of boston_3, and the result is further stacked on top of boston_4.
```python
# Load the csv files into Pandas DataFrames
boston_1 = pd.read_csv("boston_weather_1.csv")
boston_2 = pd.read_csv("boston_weather_2.csv")
boston_3 = pd.read_csv("boston_weather_3.csv")
boston_4 = pd.read_csv("boston_weather_4.csv")
more_variables = pd.read_csv("more_weather_variables.csv")

# Stack boston 1, 2, 3, and 4 vertically
combined_boston = pd.concat([boston_1, boston_2, boston_3, boston_4]).reset_index()
combined_boston = combined_boston.drop(columns=['index'])

# Horizontally merge combined_boston and more_variables
boston_data = pd.merge(combined_boston, more_variables, left_index=True, right_index=True)
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

# Part 1
Given the diabetes dataset you used during Assignment 2, implement an MLP Regressor

```python
# Load the data
data = load_diabetes()
diabetes = pd.DataFrame(data.data, columns = data.feature_names)
diabetes['progression'] = data.target
diabetes.head()
```

```python
# Get test and training data
X = diabetes.drop(['progression'], axis = 1)
y = diabetes['progression']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3)
```

```python
# Implement an MLP Regressor
mlp = MLPRegressor(hidden_layer_sizes=(50,100,50),
                       max_iter = 1000,activation = 'relu',
                       solver = 'adam')

mlp.fit(X_train, y_train)
predict = mlp.predict(X_test)
print("RMSE of MLP Regressor is: ", str(np.sqrt(mean_squared_error(predict, y_test))))
```

# Part 2
Given the prepared Titanic dataset from Assignment 3, implement a Keras sequential classifier with relu activation functions.

```python
# Get the titanic data
titanic = pd.read_csv("train.csv")
titanic.head()
```

```python
# Clean the data just like assignment 3

def impute_age(data):
    """
    Note that assigned values are found from part 1 pivot table
    """
    
    Age = data[0]
    Pclass = data[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
    
titanic['Age'] = titanic[['Age', 'Pclass']].apply(impute_age,axis=1)
titanic['Sex'].replace(['male', 'female'], [1, 0], inplace=True)
titanic['Embarked'].replace(['C', 'S', 'Q'], [0,1,2], inplace=True)
titanic = titanic.dropna(axis = 0)
titanic = titanic.drop(['Name', 'Ticket', 'PassengerId', 'Cabin'], axis = 1)
titanic = titanic.astype({"Pclass":'category', "Sex":'category', "Embarked":'category'})

X = titanic.drop(['Survived'], axis = 1)
y = titanic['Survived']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3)
```

```python
# implement a Keras sequential classifier with relu activation functions
model = Sequential()
model.add(Dense(units=7, activation='relu', input_dim=7))
model.add(Dense(activation="relu", units=100, kernel_initializer="uniform"))
model.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))

model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['accuracy']) 

model.fit(X_train, y_train, epochs=50, batch_size=30)
```

```python
# Get results
titanic_predictions = model.predict(X_test)
titanic_predictions = [1 if i > .5 else 0 for i in titanic_predictions]
print("Confusion Matrix:", '\n' + str(confusion_matrix(y_test, titanic_predictions)))
print("Accuracy is:" + str(accuracy_score(y_test, titanic_predictions)))
```

# Part 3
Given the prepared bike-share dataset from Assignment 4, implement a Keras sequential regressor with relu activation functions.

```python
# Get the data
bike = pd.read_csv("bike_share_hour.csv")
bike.head()
```

```python
# clean the data similar to assignment 4
cols = ['instant', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
temp = bike[cols]
scaler = StandardScaler()
scaler.fit(temp)
temp = scaler.transform(temp)
bike = pd.concat([bike.drop(cols, axis = 1), pd.DataFrame(temp, columns = cols)], axis=1)
bike = bike.drop(['casual', 'registered', 'dteday', 'instant'], axis = 1)

# Split the data
X = bike.drop(['cnt'], axis = 1)
y = bike['cnt']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .33)
```

```python
# Implement a Keras sequential regressor with relu activation functions
model = Sequential()
model.add(Dense(units=12, activation='relu', input_dim=12))
model.add(Dense(activation="relu", units=100, kernel_initializer="uniform"))
model.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))

model.compile('adam', loss='mean_absolute_error') 

model.fit(X_train, y_train, epochs=50, batch_size=30)
```

```python
bike_predictions = model.predict(X_test)
print("RMSE of bike data is :", str(np.sqrt(mean_squared_error(bike_predictions, y_test))))
```

# Part 4
Tune your Keras regressor from Part 3 by implementing a grid search with different optimizers.

```python
def algorithm_pipeline(X_train_data, X_test_data, y_train_data, y_test_data, 
                       model, param_grid, cv=10, scoring_fit='neg_mean_squared_error',
                       do_probabilities = False):
    gs = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, n_jobs=-1, scoring=scoring_fit, verbose=2
)
    fitted_model = gs.fit(X_train_data, y_train_data)
    
    if do_probabilities:
        pred = fitted_model.predict_proba(X_test_data)
    else:
        pred = fitted_model.predict(X_test_data)
    
    return fitted_model, pred

def build_cnn(activation = 'relu', optimizer = 'Adam'):

    model = Sequential()
    model.add(Dense(units=12, activation=activation, input_dim=12))
    model.add(Dense(activation=activation, units=100, kernel_initializer="uniform"))
    model.add(Dense(activation=activation, units=1, kernel_initializer="uniform"))
    model.compile('adam', loss='mean_absolute_error') # compile is necessary before fitting
    
    
    return model

param_grid = {'epochs':[1,3,10,20,50],
              'batch_size':[10,30,50]}

model = KerasClassifier(build_fn = build_cnn, verbose=0)

model, pred = algorithm_pipeline(X_train, X_test, y_train, y_test, model, 
                                        param_grid, cv=5, scoring_fit='neg_mean_squared_error')

print(model.best_score_)
print(model.best_params_)
```
